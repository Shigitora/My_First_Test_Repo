```python
input_matrix = input_df.values
B = input_matrix 
print('размерность матрицы', B.shape)
print()
```

```python
A = np.zeros(1000)
```

```python
C = np.zeros(1000)
```

```python
for i in range(1,999):
  A[i]= (B[i-1] + B[i] + B[i+1])/3
  

```

```python
for i in range(1,999):
    C[i]= (A[i-1] + A[i] + A[i+1])/3
```

```python
C[999] = B[999]
C[0] = B[0]
print(C)
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook

output_notebook()
```

```python
M = np.zeros(1000)
```

```python
N = 0
for i in range (1,999):
  if C[i] < 34.5:
    N = N + 1
    if N >= 5:
      M[i] = 1
  else:
    M[i] = 0
    N = 0
```

```python
print (M)
```

```python
p = figure(plot_width = 1500, plot_height = 600)

p.line(np.arange(len(input_df)),input_df , line_width = 2)
p.line(np.arange(len(M)), M, line_width = 2, color = 'red')
p.line(np.arange(len(A)), A, line_width = 2, color = 'blue')



show(p)
```
